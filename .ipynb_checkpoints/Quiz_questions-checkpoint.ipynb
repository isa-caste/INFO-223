```python
# Question 11
while True:
    player1 = input("Player 1, enter your choice (rock, paper, or scissors): ")
    player2 = input("Player 2, enter your choice (rock, paper, or scissors): ")
    if player1 == player2:
        print("It's a tie!")
    elif (player1 == "rock" and player2 == "scissors") or (player1 == "scissors" and player2 == "paper") or (player1 == "paper" and player2 == "rock"):
        print("Player 1 wins!")
    else:
        print("Player 2 wins!")
    play_again = input("Wanna play again? (yes/no): ")
    if play_again.lower() != "yes":
        break
```

```python
# Question 12
numbers = []
while True:
    num = input("Enter a number (or 'done' to be finished): ")
    if num == 'done':
        break
    numbers.append(float(num))
if numbers:
    print("Max:", max(numbers))
    print("Min:", min(numbers))
    print("User's input:", ', '.join(map(str, sorted(numbers))))
else:
    print("No numbers were entered.")
```

```python
# Question 13
def histogram(string):
    hist = {}
    for char in string:
        if char in hist:
            hist[char] += 1
        else:
            hist[char] = 1
    return hist

def print_hist_asc(histogram):
    sorted_hist = sorted(histogram.items(), key=lambda x: x[1])
    for char, frequency in sorted_hist:
        print(char, frequency)
```
