```python
# Question 1
def computePay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * rate * 1.5
        pay = regular_pay + overtime_pay
    return pay
hours_input = float(input("Enter hours worked: "))
rate_input = float(input("Enter hourly rate: "))
total_pay = computePay(hours_input, rate_input)
print("Total pay:", total_pay)
```

```python
# Question 2
def computeGrade(score):
    if score >= 0.9:
        grade = "A"
    elif score >= 0.8:
        grade = "B"
    elif score >= 0.7:
        grade = "C"
    elif score >= 0.6:
        grade = "D"
    else:
        grade = "F"
    return grade
score_input = float(input("Enter score: "))
result_grade = computeGrade(score_input)
print("Grade:", result_grade)
```

```python
# Question 3
def CelToFah(degree):
    fahrenheit = (degree * 9/5) + 32
    return fahrenheit
try:
    celsius_input = float(input("Enter temperature in Celsius: "))
    fahrenheit_output = CelToFah(celsius_input)
    print("Temperature in Fahrenheit:", fahrenheit_output)
except ValueError:
    print("Invalid input. Please enter a numeric value for Celsius temperature.")
```

```python
# Question 4
total = 0
count = 0
while True:
    number_input = input("Enter a number (or 'done' to finish): ")
    if number_input == "done":
        break
    try:
        number = float(number_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input. Please enter a numeric value or 'done'.")
if count > 0:
    average = total / count
    print("Total:", total)
    print("Count:", count)
    print("Average:", average)
else:
    print("No numbers entered.")
```

```python
# Question 5
def sum_1k(M):
    s = 0
    for k in range(1, M + 1):
        s += 1 / k
    return s
def test_sum_1k():
    expected_result = 1.8333333333333333
    computed_result = sum_1k(3)
    success = abs(expected_result - computed_result) < 1e-10
    assert success, "Test failed. Expected: {}, Computed: {}".format(expected_result, computed_result)
test_sum_1k()
print("Test passed.")
```

```python
# Question 6
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius
print("Fahrenheit\tCelsius")
for fahrenheit in range(0, 101, 10):
    celsius = fahrenheit_to_celsius(fahrenheit)
    print(fahrenheit, "\t\t", celsius)
```
