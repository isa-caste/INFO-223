```python
# Question 2
def count_strings(strings):
    count = 0
    for string in strings:
        if len(string) >= 3 and string[0] == string[-1]:
            count += 1
    return count
input_list = ['adfg', 'op', 'aba', '6226']
result = count_strings(input_list)
print("Number of strings that meet the criteria:", result)
```

```python
# Question 4
filename = input("Enter a file name: ")
try:
    file = open(filename, 'r')
except FileNotFoundError:
    print("File not found.")
    exit()
sender_counts = {}
for line in file:
    if line.startswith('From '):
        words = line.split()
        sender = words[1]
        if sender in sender_counts:
            sender_counts[sender] += 1
        else:
            sender_counts[sender] = 1
file.close()
max_count = 0
max_sender = None
for sender, count in sender_counts.items():
    if count > max_count:
        max_count = count
        max_sender = sender

print(max_sender, max_count)
```

```python
# Question 5
filename = input("Enter a file name: ")
try:
    file = open(filename, 'r')
except FileNotFoundError:
    print("File not found.")
    exit()
hour_counts = {}
for line in file:
    if line.startswith('From '):
        words = line.split()
        time = words[5]
        hour = time.split(':')[0]
        if hour in hour_counts:
            hour_counts[hour] += 1
        else:
           hour_counts[hour] = 1
file.close()
sorted_counts = sorted(hour_counts.items())

for hour, count in sorted_counts:
    print(hour, count)
```
